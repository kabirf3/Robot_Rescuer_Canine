```python
# Libraries that are needed
import torch
import cv2
import time
```

```python
# Connecting to the camera and testing video feed

# Using default camera [1]
cap = cv2.VideoCapture(0)
model = torch.hub.load('ultralytics/yolov5', 'yolov5s')

while True:
    ret, frame = cap.read()
    
    if not ret:
        break
    
    results = model(frame)
    
    cv2.imshow('Obejct Detection', results.render()[0])
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        print("Program as ended")
        break

cap.release()
cv2.destroyAllWindows
cv2.waitKey(1)
```
